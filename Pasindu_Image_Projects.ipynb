```python
from google.colab import drive
drive.mount('/content/drive')

import cv2 as cv
import matplotlib.pyplot as plt

# Read the image
img = cv.imread('/content/drive/MyDrive/angelomathews.jpg', cv.IMREAD_COLOR)
assert img is not None, "Image not found"

# Convert BGR  RGB for correct display
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Show original image
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Original Image")
plt.show()
# Subsample by factor of 2
subsampled = img_rgb[::2, ::2]  # every 2nd row and column

# Show subsampled image
plt.imshow(subsampled)
plt.axis('off')
plt.title("Subsampled Image")
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('/content/drive/MyDrive/angelomathews.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Resize using bilinear interpolation
bilinear_img = cv.resize(img_rgb, (img.shape[1]//2, img.shape[0]//2), interpolation=cv.INTER_LINEAR)

plt.imshow(bilinear_img)
plt.axis('off')
plt.title("Bilinear Interpolation")
plt.show()
```

```python
# Resize using cubic interpolation
cubic_img = cv.resize(img_rgb, (img.shape[1]*2, img.shape[0]*2), interpolation=cv.INTER_CUBIC)

plt.imshow(cubic_img)
plt.axis('off')
plt.title("Cubic Interpolation")
plt.show()
```
